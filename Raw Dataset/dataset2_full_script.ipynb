```python
import pandas as pd
```

```python
# Reading in csv
data = pd.read_csv("dataset2_full.txt", sep="\s+", skip_blank_lines=True, header=None)
```

```python
# Taking first 3 rows (relevant header details) into a separate variable
header_info = data.iloc[:3, :]
```

```python
# Dropping the header details to get the raw input data
data = data.drop([0, 1, 2])
```

```python
# Dropping columns that have NaN
data = data.dropna(axis=1)
```

```python
def get_headers():
    header_with_co = list(header_info.iloc[1, :])
    header_without_co = []
    for i in header_with_co:
        try:
            float(i)
        except:
            header_without_co.append(i)
            continue
    return header_with_co, header_without_co
header_with_coordinates, header_without_coordinates = get_headers()
```

```python
data.columns = header_without_coordinates
```

```python
def generate_coordinate_cols():
    for i in header_with_coordinates:
        if i in data.columns:
            head = "lon"
            head_col = i
        else:
            data[f"{head_col}_{head}"] = float(i)
            head = "lat"
    return data
updated_data = generate_coordinate_cols()
```

```python
updated_data["days"] = updated_data["DD-HH"].apply(lambda x: x.split("-")[0])
updated_data["hours"] = updated_data["DD-HH"].apply(lambda x: x.split("-")[1])
```
